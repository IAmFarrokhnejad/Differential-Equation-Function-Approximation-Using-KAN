```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from kan import MultKAN

# V1.4: Loop through different grid and architecture definitions overnight

def train_model_with_physics(model, epochs, optimizer, t_train, ode_residual, ic_loss_fn, patience=50):
    model.train()

    best_loss = float('inf')
    counter = 0
    losses = []

    for epoch in range(epochs):
        def closure():
            optimizer.zero_grad()
            y_pred = model(t_train)
            residual_loss = torch.mean(ode_residual(t_train, y_pred)**2)
            ic_loss = ic_loss_fn(model)
            total_loss = residual_loss + ic_loss
            total_loss.backward()
            return total_loss

        loss = optimizer.step(closure)

        if loss.item() < best_loss:
            best_loss = loss.item()
            counter = 0
        else:
            counter += 1
        if counter >= patience:
            print(f"Early stopping at epoch {epoch}")
            break

        # Log and store loss
        if epoch % 10 == 0:
            print(f"Epoch {epoch}/{epochs}, Loss: {loss.item()}")
        losses.append(loss.item())

    # Plot loss curve
    plt.plot(losses)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training Loss Curve')
    plt.grid(True)
    plt.show()

    return model

def solve_example1(architecture):
    # Define model
    model = MultKAN(width=architecture, grid=25, k=3, device='cpu')

    # Define ODE components
    def ode_residual(t, y):
        dy_dt = torch.autograd.grad(y, t, grad_outputs=torch.ones_like(y), create_graph=True)[0]
        f_term = (t +((1 + 3 * t**2) / (1 + t + t**3))) * y
        g_term = 2 * t + t**3 + t**2 * ((1 + 3 * t**2) / (1 + t + t**3))
        return dy_dt + f_term - g_term

    def ic_loss_fn(model):
        t_bc = torch.tensor([[0.0]], requires_grad=True)
        y_bc = torch.tensor([[1.0]])
        y_pred = model(t_bc)
        return torch.mean((y_pred - y_bc)**2)

    # Training data
    t_train = torch.linspace(0, 1, 200).reshape(-1, 1).requires_grad_()

    # Define optimizer
    optimizer = torch.optim.LBFGS(model.parameters(), lr=0.01, max_iter=20, tolerance_grad=1e-7, tolerance_change=1e-9, history_size=50)

    # Train model
    model = train_model_with_physics(model, epochs=100, optimizer=optimizer, t_train=t_train, ode_residual=ode_residual, ic_loss_fn=ic_loss_fn)

    # Evaluate and plot
    t_eval = torch.linspace(0, 1, 200).reshape(-1, 1).requires_grad_()
    y_pred = model(t_eval).detach().numpy()
    t_eval_np = t_eval.detach().numpy()
    y_exact_eval = (np.exp(-t_eval_np**2 / 2) + t_eval_np**5 + t_eval_np**3 + t_eval_np**2) / (t_eval_np**3 + t_eval_np + 1)

    return t_eval_np, y_pred, y_exact_eval

def solve_example2(architecture):
    # Define model
    model = MultKAN(width=architecture, grid=25, k=3, device='cpu')

    # Define ODE components
    def ode_residual(t, y):
        dy_dt = torch.autograd.grad(y, t, grad_outputs=torch.ones_like(y), create_graph=True)[0]
        f_term = 2 * y
        g_term = torch.cos(4 * t)
        return dy_dt + f_term - g_term

    def ic_loss_fn(model):
        t_bc = torch.tensor([[0.0]], requires_grad=True)
        y_bc = torch.tensor([[3.0]])
        y_pred = model(t_bc)
        return torch.mean((y_pred - y_bc)**2)

    # Training data
    t_train = torch.linspace(0, 3, 300).reshape(-1, 1).requires_grad_()

    # Define optimizer
    optimizer = torch.optim.LBFGS(model.parameters(), lr=0.01, max_iter=20, tolerance_grad=1e-7, tolerance_change=1e-9, history_size=50)

    # Train model
    model = train_model_with_physics(model, epochs=100, optimizer=optimizer, t_train=t_train, ode_residual=ode_residual, ic_loss_fn=ic_loss_fn)

    # Evaluate and plot
    t_eval = torch.linspace(0, 3, 300).reshape(-1, 1).requires_grad_()
    y_pred = model(t_eval).detach().numpy()
    t_eval_np = t_eval.detach().numpy()
    y_exact_eval = (np.sin(4 * t_eval_np) / 5 + np.cos(4 * t_eval_np) / 10 + 2.9 * np.exp(-2 * t_eval_np))

    return t_eval_np, y_pred, y_exact_eval

def solve_example3(architecture):
    # Define model
    model = MultKAN(width=architecture, grid=25, k=3, device='cpu')

    # Define ODE components
    def ode_residual(t, y):
        dy_dt = torch.autograd.grad(y, t, grad_outputs=torch.ones_like(y), create_graph=True)[0]
        f_term = y - t**2 + 1
        g_term = 0
        return dy_dt - (f_term - g_term)

    def ic_loss_fn(model):
        t_bc = torch.tensor([[0.0]], requires_grad=True)
        y_bc = torch.tensor([[0.5]])
        y_pred = model(t_bc)
        return torch.mean((y_pred - y_bc)**2)

    # Training data
    t_train = torch.linspace(0, 2, 200).reshape(-1, 1).requires_grad_()

    # Define optimizer
    optimizer = torch.optim.LBFGS(model.parameters(), lr=0.01, max_iter=20, tolerance_grad=1e-7, tolerance_change=1e-9, history_size=50)

    # Train model
    model = train_model_with_physics(model, epochs=100, optimizer=optimizer, t_train=t_train, ode_residual=ode_residual, ic_loss_fn=ic_loss_fn)

    # Evaluate and plot
    t_eval = torch.linspace(0, 2, 200).reshape(-1, 1).requires_grad_()
    y_pred = model(t_eval).detach().numpy()
    t_eval_np = t_eval.detach().numpy()
    y_exact_eval = (t_eval_np + 1)**2 - 0.5 * np.exp(t_eval_np)

    return t_eval_np, y_pred, y_exact_eval

def plot_results(t1, y1, y1_exact, t2, y2, y2_exact, t3, y3, y3_exact, architecture):
    print("Results for architecture: ", architecture)
    plt.figure(figsize=(18, 5))

    # Plot Example 1
    plt.subplot(1, 3, 1)
    plt.plot(t1, y1, 'b-', label='KAN Solution')
    plt.plot(t1, y1_exact, 'g--', label='Exact Solution')
    plt.title('Example 1: Solution')
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.grid(True)
    plt.legend()

    # Plot Example 2
    plt.subplot(1, 3, 2)
    plt.plot(t2, y2, 'r-', label='KAN Solution')
    plt.plot(t2, y2_exact, 'g--', label='Exact Solution')
    plt.title('Example 2: Solution')
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.grid(True)
    plt.legend()

    # Plot Example 3
    plt.subplot(1, 3, 3)
    plt.plot(t3, y3, 'm-', label='KAN Solution')
    plt.plot(t3, y3_exact, 'c--', label='Exact Solution')
    plt.title('Example 3: Solution')
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

def main():
    architectures = [
        [1, 20, 1],
        [1, 20, 10, 1],
        [1, 20, 10, 10, 1],
        [1, [20, 10], 1],
        [1, [20, 10], 10, 1],
        [1, [20, 10], 20, 1],
        [1, [20, 10], [10, 20], 1],
        [1, [20, 10, 10], 10, 1],
        [1, [20, 10, 10], 20, 1],
        [1, [20, 10, 10], [20, 10], 1],
        [1, [20, 10, 10], [10, 20], 1],
        [1, [20, 10, 10], [10, 20, 10], [10, 10, 20], 1],
    ]
    for architecture in architectures:
        print("Solving Example 1...")
        t1, y1, y1_exact = solve_example1(architecture)

        print("Solving Example 2...")
        t2, y2, y2_exact = solve_example2(architecture)

        print("Solving Example 3...")
        t3, y3, y3_exact = solve_example3(architecture)

        print("Plotting results...")
        plot_results(t1, y1, y1_exact, t2, y2, y2_exact, t3, y3, y3_exact, architecture)


if __name__ == "__main__":
    main()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from kan import MultKAN


def train_model(model, epochs, optimizer, t_train, ode_function, trial_solution_fn, patience=50):
    """
    Train the model using the error formulation from the referenced paper.
    """
    print("log")
    model.train()

    best_loss = float('inf')
    counter = 0
    losses = []

    for epoch in range(epochs):
        def closure():
            optimizer.zero_grad()

            # Compute trial solution and its derivative
            t_train.requires_grad_()
            y_trial = trial_solution_fn(t_train, model(t_train))
            dy_trial_dt = torch.autograd.grad(y_trial, t_train, grad_outputs=torch.ones_like(y_trial), create_graph=True)[0]

            # Compute the residual error
            residual = dy_trial_dt - ode_function(t_train, y_trial) # Trial equation - function
            loss = torch.mean(residual**2)  # Mean squared error of the residual

            loss.backward()
            return loss

        loss = optimizer.step(closure)

        if loss.item() < best_loss:
            best_loss = loss.item()
            counter = 0
        else:
            counter += 1

        if counter >= patience:
            print(f"Early stopping at epoch {epoch}")
            break

        # Log and store loss
        if epoch % 5 == 0:
            print(f"Epoch {epoch}/{epochs}, Loss: {loss.item()}")
        losses.append(loss.item())

    # Plot loss curve
    plt.plot(losses)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training Loss Curve')
    plt.grid(True)
    plt.show()

    return model


def solve_example1():
    """
    Solve example 1.
    """
    # Define model
    model = MultKAN(width=[1, 12, 1], grid=25, k=3, device='cpu')

    # Define ODE function
    def ode_function(t, y):
        f_term = (t + ((1 + 3 * t**2) / (1 + t + t**3))) * y
        g_term = 2 * t + t**3 + t**2 * ((1 + 3 * t**2) / (1 + t + t**3))
        return f_term - g_term

    # Define trial solution function
    def trial_solution_fn(t, z):
        A = 1.0  # Initial condition y(0) = 1
        return A + t * z

    # Training data
    t_train = torch.linspace(0, 1, 100).reshape(-1, 1).requires_grad_()

    # Define optimizer
    optimizer = torch.optim.LBFGS(model.parameters(), lr=0.01, max_iter=20, tolerance_grad=1e-7, tolerance_change=1e-9, history_size=50)

    # Train model
    model = train_model(model, epochs=100, optimizer=optimizer, t_train=t_train, ode_function=ode_function, trial_solution_fn=trial_solution_fn)

    # Evaluate and plot
    t_eval = torch.linspace(0, 1, 100).reshape(-1, 1).requires_grad_()
    z_pred = model(t_eval);
    y_pred = trial_solution_fn(t_eval, z_pred).detach().numpy()
    t_eval_np = t_eval.detach().numpy()
    y_exact_eval = (np.exp(-t_eval_np**2 / 2) + t_eval_np**5 + t_eval_np**3 + t_eval_np**2) / (t_eval_np**3 + t_eval_np + 1)

    plt.plot(t_eval_np, y_pred, label='Predicted Solution')
    plt.plot(t_eval_np, y_exact_eval, label='Exact Solution', linestyle='--')
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.title('Example 1')
    plt.legend()
    plt.grid(True)
    plt.show()

    return t_eval_np, y_pred, y_exact_eval


def solve_example2():
    """
    Solve example 2.
    """
    # Define model
    model = MultKAN(width=[1, 12, 1], grid=25, k=4, device='cpu')

    # Define ODE function
    def ode_function(t, y):
        f_term = 2 * y
        g_term = torch.cos(4 * t)
        return f_term - g_term

    # Define trial solution function
    def trial_solution_fn(t, z):
        A = 3.0  # Initial condition y(0) = 3
        return A + t * z

    # Training data
    t_train = torch.linspace(0, 3, 300).reshape(-1, 1).requires_grad_()

    # Define optimizer
    optimizer = torch.optim.LBFGS(model.parameters(), lr=0.01, max_iter=20, tolerance_grad=1e-7, tolerance_change=1e-9, history_size=50)

    # Train model
    model = train_model(model, epochs=100, optimizer=optimizer, t_train=t_train, ode_function=ode_function, trial_solution_fn=trial_solution_fn)

    # Evaluate and plot
    t_eval = torch.linspace(0, 3, 300).reshape(-1, 1).requires_grad_()
    z_pred = model(t_eval)
    y_pred = trial_solution_fn(t_eval, z_pred).detach().numpy()
    t_eval_np = t_eval.detach().numpy()
    y_exact_eval = (np.sin(4 * t_eval_np) / 5 + np.cos(4 * t_eval_np) / 10 + 2.9 * np.exp(-2 * t_eval_np))

    plt.plot(t_eval_np, y_pred, label='Predicted Solution')
    plt.plot(t_eval_np, y_exact_eval, label='Exact Solution', linestyle='--')
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.title('Example 2')
    plt.legend()
    plt.grid(True)
    plt.show()

    return t_eval_np, y_pred, y_exact_eval


def solve_example3():
    """
    Solve example 3 ODE
    """
    # Define model
    model = MultKAN(width=[1, 12, 1], grid=15, k=4, device='cpu')

    # Define ODE function
    def ode_function(t, y):
        f_term = y - t**2 + 1
        g_term = 0
        return f_term - g_term

    # Define trial solution function
    def trial_solution_fn(t, z):
        A = 0.5  # Initial condition y(0) = 0.5
        return A + t * z

    # Training data
    t_train = torch.linspace(0, 2, 200).reshape(-1, 1).requires_grad_()

    # Define optimizer
    optimizer = torch.optim.LBFGS(model.parameters(), lr=0.01, max_iter=20, tolerance_grad=1e-7, tolerance_change=1e-9, history_size=50)

    # Train model
    model = train_model(model, epochs=100, optimizer=optimizer, t_train=t_train, ode_function=ode_function, trial_solution_fn=trial_solution_fn)

    # Evaluate and plot
    t_eval = torch.linspace(0, 2, 200).reshape(-1, 1).requires_grad_()
    z_pred = model(t_eval)
    y_pred = trial_solution_fn(t_eval, z_pred).detach().numpy()
    t_eval_np = t_eval.detach().numpy()
    y_exact_eval = (t_eval_np + 1)**2 - 0.5 * np.exp(t_eval_np)

    plt.plot(t_eval_np, y_pred, label='Predicted Solution')
    plt.plot(t_eval_np, y_exact_eval, label='Exact Solution', linestyle='--')
    plt.xlabel('t')
    plt.ylabel('y(t)')
    plt.title('Example 3')
    plt.legend()
    plt.grid(True)
    plt.show()

    return t_eval_np, y_pred, y_exact_eval

def main():
    print("Solving example 1...")
    t1, y1_pred, y1_exact = solve_example1()

    print("Solving example 2...")
    t2, y2_pred, y2_exact = solve_example2()

    print("Solving example 3...")
    t3, y3_pred, y3_exact = solve_example3()


if __name__ == "__main__":
    main()
```
